```python
def pivot_satisfaction (airplane_train, target): 
    
    df_pivot = pd.pivot_table(
        airplane_train[["id", target, "satisfaction"]],
        index = [target], 
        columns = ["satisfaction"], 
        aggfunc = "count").reset_index()
    
    df_pivot.columns = [target, "neutral or dissatisfied", "satisfied"]
    
    df_pivot["total"] = df_pivot["satisfied"]+df_pivot["neutral or dissatisfied"]
    df_pivot["dissatisfied_rate"] = df_pivot["neutral or dissatisfied"]/df_pivot["total"]*100
    
    return df_pivot
```

```python
from sklearn.preprocessing import StandardScaler
airplane_train[['Age','Flight Distance','Departure Delay in Minutes','Arrival Delay in Minutes']]=StandardScaler().fit_transform(airplane_train[['Age','Flight Distance','Departure Delay in Minutes','Arrival Delay in Minutes']])
airplane_test[['Age','Flight Distance','Departure Delay in Minutes','Arrival Delay in Minutes']]=StandardScaler().fit_transform(airplane_test[['Age','Flight Distance','Departure Delay in Minutes','Arrival Delay in Minutes']])
```

```python
# Encoding
airplane_train['Customer Type'] = airplane_train['Customer Type'].replace(['Loyal Customer', 'disloyal Customer'], [1, 0])
airplane_train['Gender'] = airplane_train['Gender'].replace(['Male', 'Female'], [1, 0])
airplane_train['Type of Travel'] = airplane_train['Type of Travel'].replace(['Personal Travel', 'Business travel'], [0, 1])
airplane_train['Class'] = airplane_train['Class'].replace(['Eco', 'Eco Plus', 'Business'], [0, 1, 2])

airplane_test['Customer Type'] = airplane_test['Customer Type'].replace(['Loyal Customer', 'disloyal Customer'], [1, 0])
airplane_test['Gender'] = airplane_test['Gender'].replace(['Male', 'Female'], [1, 0])
airplane_test['Type of Travel'] = airplane_test['Type of Travel'].replace(['Personal Travel', 'Business travel'], [0, 1])
airplane_test['Class'] = airplane_test['Class'].replace(['Eco', 'Eco Plus', 'Business'], [0, 1, 2])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, plot_roc_curve

lr_clf=LogisticRegression()
svm_clf=SVC()
dt_clf=DecisionTreeClassifier()
rf_clf=RandomForestClassifier()
```
